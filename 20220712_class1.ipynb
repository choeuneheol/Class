```python
import cv2
import numpy as np
import sys 

cv2.__version__
```

```python
img = cv2.imread("fig/puppy.bmp")

# 오류를 발견하기위한 예외 처리
if img is None:
    print("Image read failed")
    sys.exit()

cv2.imshow("image",img)
cv2.waitKey()
cv2.destroyAllWindows()
```

### cv2.imread(filename. flags = None) -> retval
### flags:
### cv2.IMREAD_COLOR: BGR color로 읽기
### cv2.IMREAD_GRAYSCALE: 그레이 color로 읽기
### cv2.IMREAD_UNCHANGED: 파일 속성대로 읽기
### cv2.IMREAD_REDUCED_GRAYSCALE_2 
### cv2.IMREAD_REDUCED_COLOR_2 

### retval: numpy.ndarray로 반환

```python
img2 = cv2.imread("fig/cat.bmp", cv2.IMREAD_GRAYSCALE)

if img2 is None:
    print("Image read failed")
    sys.exit()

cv2.namedWindow("image1")
cv2.imshow("image1",img2)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img3 = cv2.imread("fig/puppy.bmp")


# 오류를 발견하기위한 예외 처리
if img3 is None:
    print("Image read failed")
    sys.exit()

img4 = cv2.resize(src=img3, dsize = (900,700))
    
cv2.imshow("image",img3)
cv2.imshow("image2",img4)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 아스키코드 27 = [ESC], 32 = [space]

img5 = cv2.imread("fig/cat.bmp", cv2.IMREAD_GRAYSCALE)

if img5 is None:
    print("Image read failed")
    sys.exit()
    
cv2.namedWindow("image", cv2.WINDOW_AUTOSIZE)
cv2.imshow("image",img5)


while True:
    if cv2.waitKey() == 32:
        break

cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("fig/cat.bmp", cv2.IMREAD_COLOR)

if img is None:
    print("image read failed")
    sys.exit()
    
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
imgBGR = cv2.imread("fig/cat.bmp", cv2.IMREAD_COLOR)

if img is None:
    print("image read failed")
    sys.exit()
    
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB) 

plt.imshow(imgRGB)
plt.axis("off")
plt.show()
```

```python
imgGray = cv2.imread("fig/cat.bmp", cv2.IMREAD_GRAYSCALE)

if img is None:
    print("image read failed")
    sys.exit()
    

plt.imshow(imgGray, cmap = "gray") #cmap 추가 해줘야 그레이스케일 표현 가능
plt.axis("off")
plt.show()
```

```python
imgBGR = cv2.imread("fig/puppy.bmp", cv2.IMREAD_COLOR)
imgGRAY = cv2.imread("fig/puppy.bmp", cv2.IMREAD_GRAYSCALE)

if imgBGR is None or imgGRAY is None:
    print("image read failed")
    sys.exit()

imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,20))
plt.subplot(131), plt.imshow(imgBGR), plt.axis("off")
plt.subplot(132), plt.imshow(imgRGB), plt.axis("off")
plt.subplot(133), plt.imshow(imgGRAY, cmap = "gray"), plt.axis("off")
plt.show()
```
