```python
img_files = glob.glob("./fig1/images/*.*")

cv2.namedWindow("scene", cv2.WINDOW_NORMAL)
cv2.setWindowProperty("scene", cv2.WND_PROP_FULLSCREEN,cv2.WINDOW_FULLSCREEN)

idx = 0

while True:
    img = cv2.imread(img_files[idx])
    
    if img is None:
        print("image read failed")
        break
        
    cv2.imshow("scene", img)
    
    if cv2.waitKey(2000) == 27:
        break
    
    idx += 1
    
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
fourcc = cv2.VideoWriter_fourcc(*"DIVX")

out = cv2.VideoWriter("output_edge.avi", fourcc = fourcc, fps = fps, frameSize =(w,h))

if not cap.isOpened():
    print("Cap open failed")
    sys.exit()

cv2.namedWindow("frame")

while True:
    ret, frame = cap.read()
    
    if not ret:
        print("Go home!")
        break
        
    out.write(frame)
    cv2.imshow("frame", frame)
    
    if cv2.waitKey(30) == 27:
        break

out.release()
cap.release()
cv2.destroyAllWindows()
```

```python
img = cv2.imread("fig1/cow.png",0)

cv2.imshow("img",img)

while True:
    key = cv2.waitKey()
    
    if key ==27 or key == ord("q"):
        break
        
    elif key == ord("i"):
        img2 = 255-img
        cv2.imshow("img",img2)
        
    elif key == ord("e"):
        img3 = cv2.Canny(img, 50, 150)
        cv2.imshow("img", img3)
    
    elif key == ord("r"):
        cv2.imshow("img",img)
        
cv2.destroyAllWindows()
```

```python
# 키보드 처리
img = cv2.imread("./fig1/cat.bmp", cv2.IMREAD_GRAYSCALE)

if img is None:
    print("image read failed")
    sys.exit()
    
cv2.imshow("image", img)

while True:
    key = cv2.waitKey()
    if key == 27 or key == ord("q"):
        break
    
    elif key == ord("i"):
        img1 = 255 - img
        cv2.imshow("image",img1)
        
    elif key == ord("e"):
        edge = cv2.Canny(img, 50, 150)
        cv2.imshow("image", edge)
    
    elif key == ord("r"):
        cv2.imshow("image", img)
        
cv2.destroyAllWindows()
```

```python
# 영상 효과 넣기

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Video capture failed")
    sys.exit()

f_flag = False
i_flag = False

while True:
    
    ret, frame = cap.read()
    
    if not ret:
        print("Go home!")
        break
    
    if f_flag:
        frame = cv2.flip(frame, 1)
        
    if i_flag:
        frame = cv2.Canny(frame, 50, 150)
        
    cv2.imshow("frame",frame)
    
    key = cv2.waitKey(30)
    
    if key == 27:
        break
    
    elif key == ord("f"):
        f_flag = not f_flag
        
    elif key == ord("i"):
        i_flag = not i_flag
    
    

cap.release()
cv2.destroyAllWindows()
```

```python
## 마우스 콜백

def call_mouse(event,x,y,flags,param):
    
    if event == cv2.EVENT_LBUTTONDOWN:
        print("left button down",x,y)
        
    elif event == cv2.EVENT_LBUTTONUP:
        print("left button up",x,y)
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if flags == cv2.EVENT_FLAG_CTRLKEY:
            print("Mouse move", x, y)
    



img = np.ones((600,1000,3), np.uint8)*255

cv2.namedWindow("image")

cv2.setMouseCallback("image", call_mouse, img)

cv2.imshow("image", img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
def call_mouse(event,x,y,flags,param):
    global oldx, oldy
    
    if event == cv2.EVENT_LBUTTONDOWN:
        oldx, oldy = x, y
        
#     elif event == cv2.EVENT_LBUTTONUP:
#         print("left button up",x,y)
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if flags == cv2.EVENT_LBUTTONDOWN:
            cv2.line(img,(oldx, oldy),(x,y),(0,0,255),8,cv2.LINE_AA)
            cv2.imshow("image", img)
            oldx, oldy = x, y
            
    



img = np.ones((600,1000,3), np.uint8)*255

cv2.namedWindow("image")

cv2.setMouseCallback("image", call_mouse, img)

cv2.imshow("image", img)

cv2.waitKey()
cv2.destroyAllWindows()
```
