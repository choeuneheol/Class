```python
img_files = glob.glob("./fig/images/*.*")
print(img_files)

cv2.namedWindow("image", cv2.WINDOW_NORMAL)
cv2.setWindowProperty("image", cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)


idx = 0

while True:
    img = cv2.imread(img_files[idx])
    
    if img is None:
        print("image read failed")
        break
        
    if idx >= len(img_files):
        print()
        
cv2.imshow('img_files', img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
src = cv2.imread("./fig/airplane.bmp")
mask = cv2.imread("./fig/mask_plane.bmp", cv2.IMREAD_GRAYSCALE)
dst = cv2.imread("./fig/field.bmp", cv2.IMREAD_COLOR)

cv2.copyTo(src, mask, dst)

cv2.imshow("src", src)
cv2.imshow("mask", mask)
cv2.imshow("dst", dst)

cv2.waitKey()

cv2.destroyAllWindows()
```

```python
src = cv2.imread("./fig/cow.png")
dst = cv2.imread("./fig/green.png")

src_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

ret, mask = cv2.threshold(src_gray, 240, 255, cv2.THRESH_BINARY_INV)


h, w = src.shape[:2]

dst_crop = dst[400:400 + h, 300:300+w]

cv2.copyTo(src, mask,dst_crop)



cv2.imshow("src", src)
cv2.imshow("dst", dst)
cv2.imshow("mask", mask)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
src = cv2.imread("./fig/puppy.bmp")
sunglass = cv2.imread("./fig/imgbin_sunglasses_1.png", cv2.IMREAD_UNCHANGED)
sunglass = cv2.resize(sunglass, (300, 150))

mask = sunglass[:,:,-1]
glass = sunglass[:,:,0:3]

h, w = mask.shape[:2]

src_crop = src[120:120+h, 220:220+w]

cv2.copyTo(glass, mask, src_crop)

cv2.imshow("src", src)
cv2.imshow("mask", mask)
cv2.imshow("glass", glass)
cv2.imshow("crop", src_crop)


cv2.waitKey()
cv2.destroyAllWindows()
```
